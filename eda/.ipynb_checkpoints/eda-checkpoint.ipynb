```python
import os 
import json
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from tqdm import tqdm
```

```python
folder_path = "result_jsons/"
result_jsons = []
```

```python
params_dict = {}
# in billions
params_dict["2.8b"] = 2.8
params_dict["1.4b"] = 1.4
params_dict["1b"] = 1.0
params_dict["410m"] = 0.41
params_dict["160m"] = 0.16
```

```python
for json_file in tqdm(sorted(os.listdir(folder_path))):
    weak_params, weak_steps, strong_params, strong_steps = json_file.replace("EleutherAI_pythia-","").replace("step","").split("_")[:4]
    with open(os.path.join(folder_path,json_file),"r") as f:
        result = json.loads(f.read())
    result["weak_params"] = params_dict[weak_params]
    result["weak_steps"] = weak_steps
    result["strong_params"] = params_dict[strong_params]
    result["strong_steps"] = strong_steps
    result_jsons.append(result)
```

```python
results_df = pd.DataFrame(result_jsons)
```

```python
results_df = results_df.astype("float")
```

```python
results_df.head(10)
```

```python
results_df.to_csv("results_df.csv")
```

```python
results_df["weak_model"] = results_df["weak_params"].astype(str)+"_"+results_df["weak_steps"].astype(int).astype(str)
results_df["strong_model"] = results_df["strong_params"].astype(str)+"_"+results_df["strong_steps"].astype(int).astype(str)
```

```python
results_df
```

```python
unique_weak_steps = results_df['weak_steps'].unique()
unique_strong_steps = results_df['strong_steps'].unique()

# Create subplots
fig, axes = plt.subplots(len(unique_weak_steps), len(unique_strong_steps), figsize=(12, 8))

for i, weak_step in enumerate(unique_weak_steps):
    for j, strong_step in enumerate(unique_strong_steps):
        # Subset the dataframe for the current weak_step and strong_step
        subset_df = results_df[(results_df['weak_steps'] == weak_step) & (results_df['strong_steps'] == strong_step)]
        
        # Scatter plot for the subset
        sc = axes[i, j].scatter(subset_df['transfer_acc_xent'], subset_df['transfer_acc_logconf'], cmap='viridis', label='acc_ent')
        
        # Add colorbar for each subplot
        cbar = plt.colorbar(sc, ax=axes[i, j])
        cbar.set_label('Accuracy')
        
        # Set labels and title for each subplot
        axes[i, j].set_xlabel('Transfer Acc Xent')
        axes[i, j].set_ylabel('Transfer Acc Logconf')
        axes[i, j].set_title(f'Plot for weak={weak_step}, strong={strong_step}')
        
        # Show legend for the first subplot
        if i == 0 and j == 0:
            axes[i, j].legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
df = results_df

# Set the style of seaborn for better aesthetics
sns.set(style="whitegrid")

# Create subplots for each independent variable
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Plot for weak_params
sns.scatterplot(x='weak_params', y='transfer_acc_xent', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Weak Params vs Transfer Acc Xent')

# Plot for strong_params
sns.scatterplot(x='strong_params', y='transfer_acc_xent', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Strong Params vs Transfer Acc Xent')

# Plot for weak_steps
sns.scatterplot(x='weak_steps', y='transfer_acc_xent', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Weak Steps vs Transfer Acc Xent')

# Plot for strong_steps
sns.scatterplot(x='strong_steps', y='transfer_acc_xent', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Strong Steps vs Transfer Acc Xent')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
